For our analysis ,We have taken those columns which we need for our questions.

```python
cols_to_keep = ['code','product_name','packaging','packaging_tags','energy_100g', 'brands','alcohol_100g',
        'proteins_100g','fiber_100g','salt_100g','sodium_100g','sugars_100g','trans-fat_100g',
        'main_category_en','nutriscore_grade','cholesterol_100g','carbohydrates_100g','saturated-fat_100g',
        'fat_100g','categories_en','labels_tags','countries_en',
        'ingredients_text','ingredients_from_palm_oil_n','additives_n','nutriscore_score','categories_en']
```

```python
foodfact_worldwide=foodfact[cols_to_keep]
```

```python
print(foodfact_worldwide.dtypes)
```

We have decided to fill the missing nutrient values with 0's. Because it can be assumed that the header of a product only specifies the amount of nutrients present in the product. 
(For example, there is no data for vegetable oil containing Suger)

Replacing float values  with 0 because we may loose data if we will do dropna() in Questions. 

```python
column_to_replacezero=['alcohol_100g', 'proteins_100g', 'fiber_100g', 'salt_100g',
       'sodium_100g', 'sugars_100g', 'trans-fat_100g', 'cholesterol_100g',
       'carbohydrates_100g', 'saturated-fat_100g', 'fat_100g',
       'energy_100g', 'nutriscore_score','additives_n','ingredients_from_palm_oil_n']
df=foodfact_worldwide
foodfact_worldwide[column_to_replacezero]=foodfact_worldwide[column_to_replacezero].fillna(0)

```

```python
foodfact_worldwide.product_name = foodfact_worldwide.product_name.astype(str)
```

Renaming columns where hyphen is used in order to avoid errors in python 3

```python
foodfact_worldwide=foodfact_worldwide.rename(columns={'saturated-fat_100g':'saturated_fat_100g','trans-fat_100g':'trans_fat_100g'})
```

```python
foodfact_worldwide.to_csv('openfoodfactnew.csv',index=False)
```
